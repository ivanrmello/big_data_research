```python
import pandas as pd
from datetime import datetime
from time import gmtime, strftime
import time
import pyautogui
import pyperclip
```

```python
def get_now_time():
    now = datetime.now()
    return now.strftime("%d/%m/%Y %H:%M:%S")
```

```python
def get_app_details_data_from_play_store(app_id):

    # opening a new incognito window
    pyautogui.hotkey('ctrl', 'shift', 'n')

    # waiting 0.5 seconds
    time.sleep(0.5)

    # defining app_id
    app_id = app_id

    # copying and pasting URL and pressing enter
    pyperclip.copy('https://play.google.com/store/apps/details?id=' + app_id)
    time.sleep(0.2)
    pyautogui.hotkey('ctrl','v')
    time.sleep(0.2)
    pyautogui.press('enter')
    time.sleep(2)

    # checking if app is not available anymore
    pyautogui.hotkey('ctrl','a')
    time.sleep(1)
    pyautogui.hotkey('ctrl','c')
    time.sleep(1)
    s = pyperclip.paste()
    page_content_as_list = s.split('\r\n')
    if page_content_as_list[0]=='O URL solicitado não foi encontrado no servidor.':
        app_launch_info = 'O URL solicitado não foi encontrado no servidor.'
    else:
        app_launch_info = 'OK, existem dados.'

    # Searching for "Sobre este app"
    pyautogui.hotkey('ctrl','f')
    time.sleep(0.2)
    pyperclip.copy('Sobre este app')
    time.sleep(0.2)
    pyautogui.hotkey('ctrl','v')
    time.sleep(0.2)
    pyautogui.press('enter')
    time.sleep(0.2)
    pyautogui.press('esc')
    time.sleep(0.2)
    pyautogui.press('tab')
    time.sleep(0.2)
    pyautogui.press('enter')
    time.sleep(0.2)
    pyautogui.hotkey('ctrl','a')
    time.sleep(0.2)
    pyautogui.hotkey('ctrl','c')
    time.sleep(0.2)
    s = pyperclip.paste()

    # pressing ESC to leave pop-up, then ALT + F4 to close incognito window
    pyautogui.press('esc')
    time.sleep(0.2)
    pyautogui.hotkey('alt','f4')

    # transforming s in a list of elements
    page_content_as_list = s.split('\r\n')

    if app_launch_info == 'OK, existem dados.':
        try:
            app_launch_info = page_content_as_list[page_content_as_list.index('Lançado em') + 1]
        except:
            app_launch_info = None

        try:
            app_version_info = page_content_as_list[page_content_as_list.index('Versão') + 1]
        except:
            app_version_info = None

        try:
            app_update_info = page_content_as_list[page_content_as_list.index('Atualizado em') + 1]
        except:
            app_update_info = None

        try:
            app_downloads_info = page_content_as_list[page_content_as_list.index('Downloads') - 1]
        except:
            app_downloads_info = None

        try:
            app_content_classification = page_content_as_list[page_content_as_list.index('Classificação do conteúdo') + 1]
        except:
            app_content_classification= None

        try:
            app_rating_score = page_content_as_list[page_content_as_list.index('star') - 1]
        except:
            app_rating_score= None

        try:
            app_rating_quantity = page_content_as_list[page_content_as_list.index('star') + 1]
        except:
            app_rating_quantity= None

        # creating a dict with relevant info
        d = {'app_id':app_id,
            'app_launch_info':app_launch_info,
            'app_version_info':app_version_info,
            'app_update_info':app_update_info,
            'app_downloads_info':app_downloads_info,
            'app_content_classification':app_content_classification,
            'app_rating_score':app_rating_score,
            'app_rating_quantity':app_rating_quantity,
            'app_scraped_datetime':get_now_time()
            }
    else:
        # 'O URL solicitado não foi encontrado no servidor.'
        d = {'app_id':app_id,
            'app_launch_info':app_launch_info,
            'app_version_info':app_launch_info,
            'app_update_info':app_launch_info,
            'app_downloads_info':app_launch_info,
            'app_content_classification':app_launch_info,
            'app_rating_score':app_launch_info,
            'app_rating_quantity':app_launch_info,
            'app_scraped_datetime':get_now_time()
            }
    # parsing dict to dataframe
    df_app_info = pd.DataFrame(data=d, index=[0])

    # exporting dataframe as excel file
    df_app_info.to_excel('C:\\Users\\ivanr\\Documents\\83 - Mestrado UFG\\mobile_apps_data\\play_store_details_data\\' + app_id.replace(".","") + '.xlsx', index=False)
```

```python
def get_app_safety_data_from_play_store(app_id):

    # opening a new incognito window
    pyautogui.hotkey('ctrl', 'shift', 'n')

    # waiting 0.5 seconds
    time.sleep(0.5)

    # defining app_id
    app_id = app_id

    # copying and pasting URL and pressing enter
    pyperclip.copy('https://play.google.com/store/apps/datasafety?id=' + app_id)
    time.sleep(0.2)
    pyautogui.hotkey('ctrl','v')
    time.sleep(0.2)
    pyautogui.press('enter')
    time.sleep(2)

    # checking if app is not available anymore
    pyautogui.hotkey('ctrl','a')
    time.sleep(1)
    pyautogui.hotkey('ctrl','c')
    time.sleep(1)
    s = pyperclip.paste()
    page_content_as_list = s.split('\r\n')
    #print(page_content_as_list)
    if page_content_as_list[0]=='O URL solicitado não foi encontrado no servidor.':
        # pressing ESC to leave pop-up, then ALT + F4 to close incognito window
        pyautogui.press('esc')
        time.sleep(0.2)
        pyautogui.hotkey('alt','f4')
        app_launch_info = ['O URL solicitado não foi encontrado no servidor.']
        # 'O URL solicitado não foi encontrado no servidor.'
        df_collected_data = pd.DataFrame(columns=['app_id','app_collected_data_details','app_scraped_datetime'])
        df_collected_data['app_collected_data_details'] = app_launch_info
        df_collected_data['app_id'] = app_id
        df_collected_data['app_scraped_datetime'] = get_now_time()
        df_collected_data.to_excel('C:\\Users\\ivanr\\Documents\\83 - Mestrado UFG\\mobile_apps_data\\\play_store_safety_data\\' + app_id.replace(".","") + '.xlsx', index=False)
    else:
        app_launch_info = ['OK, existem dados.']
        pyautogui.hotkey('ctrl','a')
        time.sleep(1.5)
        pyautogui.hotkey('ctrl','c')
        time.sleep(1.5)
        s = pyperclip.paste()

        # pressing ESC to leave pop-up, then ALT + F4 to close incognito window
        pyautogui.press('esc')
        time.sleep(0.2)
        pyautogui.hotkey('alt','f4')

        # transforming s in a list of elements
        page_content_as_list = s.split('\r\n')

        if app_launch_info[0]=='OK, existem dados.':

            # getting privacy_tip info
            try:
                privacy_tip = [page_content_as_list[page_content_as_list.index('privacy_tip') + 1]]
            except:
                privacy_tip = ['Privacy Tip não foi encontrada na página']

            # testing if privacy_tip tells that developer does not share info about collected data
            if privacy_tip[0] == 'Não há informações do desenvolvedor sobre como o app coleta, processa e compartilha seus dados':
                collected_data = ['Não há informações do desenvolvedor sobre como o app coleta, processa e compartilha seus dados']
            else:
                # getting collected data info
                try:
                    collected_data = page_content_as_list[page_content_as_list.index('Dados coletados'):page_content_as_list.index('Práticas de segurança') -1]
                    indices = [i for i, x in enumerate(collected_data) if x == "expand_more"]
                    collected_data_parsed_list = []
                    for element_index in indices:
                        generic = collected_data[element_index-2]
                        detailed = collected_data[element_index-1]
                        collected_data_parsed_list.append("Generic: {} | Detailed: {}".format(generic, detailed))
                    collected_data = collected_data_parsed_list
                except:
                    collected_data = ["Apesar do desenvolvedor divulgar informações sobre como o app coleta, processa e compartilha seus dados, não foi possível coletar a seção de 'Dados Coletados'."]

            df_collected_data = pd.DataFrame(columns=['app_id','app_collected_data_details','app_scraped_datetime'])
            df_collected_data['app_collected_data_details'] = collected_data
            df_collected_data['app_id'] = app_id
            df_collected_data['app_scraped_datetime'] = get_now_time()
            df_collected_data.to_excel('C:\\Users\\ivanr\\Documents\\83 - Mestrado UFG\\mobile_apps_data\\\play_store_safety_data\\' + app_id.replace(".","") + '.xlsx', index=False)
```

```python
play_store_apps_id_list = list(df_app_ids[df_app_ids['play_store_app_id'].isna()==False]['play_store_app_id'])
```

```python
for app_id in play_store_apps_id_list[5:]:
    # waiting 5 seconds to activate chrome window
    time.sleep(1.5)
    get_app_details_data_from_play_store(app_id)
    time.sleep(1.5)
    get_app_safety_data_from_play_store(app_id)
```
